```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

```python
import matplotlib.pyplot as plt
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Line plot")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Line plot")
```

```python
sns.barplot(x="total_bill", y="tip", data=tips)
plt.title("Line plot")
```

```python
sns.boxplot(x="total_bill", y="tip", data=tips)
plt.title("Line plot")
```

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

#### Create pivot table

```python
pv = flights.pivot(index="month", columns="year", values="passengers")
```

```python
pv
```

```python
#  annot = display numbers, fmt = format "d" --> Intrger, cmap = colormap, "YlGnBu"yellow.green.blue
sns.heatmap(pv, annot=True, fmt="d", cmap="YlGnBu")
```

```python
import pandas as pd
 
df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})
 
sns.scatterplot(x="age", y="salary", hue="gender", data=df)
plt.title('Salary vs Age Scatter Plot')
plt.show()
```
