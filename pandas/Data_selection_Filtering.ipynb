### Usind iloc [index position]
#### --> The ilog means that the value of i Will be used.
#### --> The value of i Always start from 0 if you have read loops from python.

```python
df.iloc[8,5] # df.loc[8,'IMDb'] --> same as this diff only 'IMDB' instead index position 5
```

#### What is the diffrence?
--> We use labels in loc and in ilog We use indices which start from 0.

```python
df.loc[0:2 ,['Actor', 'Year','Film']] # last index as inclusive.
```

```python
df.iloc[0:2 ,[0,2,3]] # last index as exclusive.
```

### Using at and iat method

### To access a Particular element. and this is the fastest method. it gives you fast access to a particular element.

```python
df.at[0,'Film']
```

```python
df.iat[0,1] # df.at[0,'Film']--> same as this diff only 'Film' instead index position 1
```

#### Note:- The at and iat are optimized for just  access only one element.
--> it is only for element access. to access a particular cell.
