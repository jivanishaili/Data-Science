```python
# python zip explained
l1 = [1,2,3]
l2 = [4,5,6]
list(zip(l1,l2))
```

```python
# using python lists
import numpy as np
import time

size = 1_000_000_0
l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x, y in zip(l1,l2)]
end = time.time()
print(add[0:10])

print(end - start)
```

```python
# using numpy arrays
import numpy as np
import time

size = 1_000_000_0
l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
end = time.time()
print(add[0:10])

print(end - start)
```
