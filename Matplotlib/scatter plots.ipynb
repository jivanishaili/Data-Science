```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1,2,3,4,5,6,7,8,9]
exam_scores = [40,45,50,55,60,65,75,85,90]
```

```python
colors = ['purple' if score<=50 else 'pink' for score in exam_scores]
sizes = [score *2 for score in exam_scores]
plt.scatter(study_hours, exam_scores, c=colors, s = sizes )
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study VS Marks")
```

```python
# colors = ['purple' if score<=50 else 'pink' for score in exam_scores]
sizes = [score *2 for score in exam_scores]
plt.scatter(study_hours, exam_scores, s = sizes, c = exam_scores, cmap='viridis')
plt.colorbar(label="Score")
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study VS Marks")
```

```python
# colors = ['purple' if score<=50 else 'pink' for score in exam_scores]
sizes = [score *2 for score in exam_scores]
social_score = [12,34,67,555,678,145,56,999,445]
plt.scatter(study_hours, exam_scores,  c = social_score, cmap='viridis')
plt.colorbar(label="social_Score")
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study VS Marks")
```
