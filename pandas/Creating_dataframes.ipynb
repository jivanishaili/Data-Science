#### Using python [List of list] With Dataframe

```python
import pandas as pd
```

```python
data = [["Harry",89],["Shelu",78],["Devu",88],["Shrut",98],["Granthu",58]]
```

```python
pd.DataFrame(data, columns=["Name","Marks"])
```

#### Using python Dictionary  Dataframe

```python
data = {
   "A" : [21,56,54],
    "B":[87,58,95]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```
