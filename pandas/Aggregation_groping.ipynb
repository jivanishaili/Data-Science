```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Department":["HR","HR","IT","IT","Marketing","Marketing","Sales","Sales"],
    "Team":["A","A","B","B","C","C","D","D"],
    "Gender":["M","F","M","F","M","F","M","F"],
    "Salary":[85,90,78,85,92,88,75,80],
    "Age":[23,25,30,22,28,26,21,27],
    "JoinDate":pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20","2021-04-10",
        "2020-05-30", "2020-06-25" , "2021-07-15", "2021-08-01"    
    ]) 
})
```

```python
df
```

#### How you can use these common Aggregation func. with group buy

```python
df.groupby("Department")["Salary"].mean()
```

```python
type(df.groupby("Department")["Salary"].mean())
```

```python
df.groupby("Team")["Salary"].sum()
```

```python
df.groupby("Team")["Salary"].count()
```

```python
df.groupby("Department")["Salary"].min()
```

```python
df.groupby("Department")["Salary"].max()
```

```python
df.groupby("Department")["Salary"].count()
```

#### AGG and Aggregation function in pandas they are exactly same.

```python
df.groupby("Team")["Salary"].agg(["mean","max","min"])
```

#### This is not series, this is a data frame. it's output is a data frame.

```python
type(df.groupby("Team")["Salary"].agg(["mean","max","min"]))
```

#### If you want that the name of my column is of my choice, then you run something like this.
--> your column name will be of your choice.

```python
df.groupby("Team")["Salary"].aggregate(
    avg_score = "mean",
    high_score = "max"
)
```

```python
df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean")
```
