### 1. For loop

```python
color=['Red','Blue','White','Pink']
for colors in color:
    print(colors)
```

### Using Range()

```python
for i in range(1,11): #range(by defult 0 to n-1) but you define 1 so starting position take 1.
    print(i) 
```

### While loop

#### While loop is a loop that runs As long as the condition is true.

```python
count = 0
while count<4:
    print(count)
    count+=1
```

```python
i = 0
while i<12:
    print(i)
    i = i + 1
```

### Loop control statement

#### pass - Do nothing(used as a placeholder)
--> use in loop or function

```python
n=[2,3,4,5]
for no in n:
    pass # Don"t do anything
    
print("Hello!")
```

### Break statement

```python
l = ["apple","cherry","mango","banana"]
for item in l:
    if (item=="mango"):
        break # stop this loop right here
    print(item)
```

### continue statement
-> skip current iteration move to next iteration

```python
num=[3,4,5,6,7]
for i in num:
    if i == 5:
        continue # skip this particular iteration of the loop
    print(i)
```
