```python
import matplotlib.pyplot as plt 

plt.style.use("ggplot") # choose any style you like
```

```python
#data
labels = ["Sachin","Sehwag","Kohli","Yuwraj"]
runs=[18000,8000,12000,9500]

plt.title("Career runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    wedgeprops={'edgecolor':'#127806','linewidth':2, 'linestyle':'--', 'fill':False}
)
```

```python
explode = [0.1,0,0,0] # Only hightlight sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors, shadow=True, startangle=140, autopct ='%1.1f%%') # adds a 3D like shadow , startangle ->To rotate the pie chart.
plt.title("Exploded pie Example")
plt.show()
```
