### Using f-string method

```python
lan="python"
sub="Maths"
print(f"My favourite languages is {lan} and My favourite subject is {sub}.")
```

```python
a = 10
b = 55
print(f"sum of {a} and {b} is {a+b}")
```

#### Formatting Numbers

```python
pi = 3.14159
print(f"pi rounded to 2 decimal places : {pi:.2f}")
```

### Padding and Alignment

#### :<10 -> The < symbol means left-align the text within a total width of 10 characters

##### F-strings are the the most efficient and recommended way to format strings in morden python!

```python
print(f"{'Python':<10}") #Left align
print(f"{'Python':>10}") #right align
print(f"{'Python':^10}") #center align
```

```python
s = 'shaili'
print(s[0:4])
```

```python
print(f"{"Shaili_jivani":<20}")
print(f"{"Shaili_jivani":>20}")
print(f"{"Shaili_jivani":^20}")
```
