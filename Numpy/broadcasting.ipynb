```python
import numpy as np
```

```python
arr = np.array([1,4,3,5,7])
result = arr ** 2 #vectorized operation
result
```

```python
result = np.array([5,6,7,8])
result + 10 # Broadcasting : 10 is added to all elements
```

```python
arr1 = np.array([[3,2,2],[5,6,7]])
arr2 = np.array([1,4,5])
arr = arr1 + arr2 #Broadcasting arr2 across arr1
print(arr)
```

```python
data = np.array([[10,20,30],
                 [15,25,35],
                 [20,30,40],
                 [25,35,45],
                 [30,40,50]])
mean = data.mean(axis=0)
std = data.std(axis=0)

normalized_data = (data - mean) / std
print(normalized_data)
```
