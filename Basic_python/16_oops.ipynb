```python
class Employee:
    company="HP"
    
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
        
    def printDetails(self):
        print(f"The company of {self.name} is {self.company} with salary ${self.salary}")

    @staticmethod
    def printTime():
        print(f"Time start now.")

    @classmethod
    def Printclassdetials(cls):
        print(f"The company is {cls.company}.")
        
e = Employee("Harry",47000) 
e.printDetails()
e.printTime()
e.Printclassdetials()
```
