#### ->simply a python library. same as numpy and pandas.

```python
import matplotlib.pyplot as plt
```

#### line plots

```python
plt.plot([1,2,3],[4,5,6])
```

```python
years = [1990,1992,1994,1996,1998,2000,2003,2005,2007,2010]
runs = [500,700,1100,1500,1800,1200,1700,1300,900,1500]
plt.plot(years,runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's yearly Runs")
```

```python
Kohli = [0,0,500,800,1100,1300,1500,1800,1900,2100]
Sehwag = [0,300,800,1200,1500,1700,1600,1400,1000,0]
plt.plot(years,Kohli,label="Virat kohli")
plt.plot(years,Sehwag,label="Virendeer Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("performance comparision")
plt.legend()
plt.show()
```

#### Format String
#### format = '[marker][line][color]'
--> other combination such as [color][marker] [line] are also supported.

```python
plt.plot(years,Kohli,'mo--',label="Virat kohli") # mgenta color with circle marker and dashed lines
plt.plot(years,Sehwag,'g^:',label="Virendeer Sehwag") # green traingles dotted
plt.legend()
```

#### linewidth = 2 you can change the line width
#### plt.grid(True) --> you can see the grid here
Grid mins background lining

```python
plt.plot(years,Kohli,color="orange",linestyle="--",label="Virat kohli")
plt.plot(years,Sehwag,color="green",linestyle="-.",linewidth=2,label="Virendeer Sehwag")
plt.plot(years,runs,color="blue",label="Tendulkar")
plt.tight_layout()
plt.grid(True)
plt.legend()
```
