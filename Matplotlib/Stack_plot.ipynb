```python
import matplotlib.pyplot as plt
```

```python
activities = ['Styding','Playiing','Watching TV','Sleeping']
time_spent = [3,2,2,5]
colors = ["red","blue","yellow","pink"]
plt.figure(figsize=(6,6))
plt.pie(time_spent, labels=activities, colors=colors, shadow=True,startangle=90, autopct="%1.1f%%")
plt.title("School going student")
plt.axis("equal")
```

#### alpha = 0.4 mins alpha transparency level from 0 to 1 you can change the thickness to 1 if you want light then change it to 0 

```python
days = [1,2,3,4,5,6,7]
studying = [3,4,3,5,4,3,4]
playing = [2,2,1,1,2,3,2]
Watching_TV = [2,1,2,2,1,1,1]
sleeping = [5,5,6,5,6,5,5]

labels = ["Studying","playing","TV","sleeping"]
colors = ["red","blue","yellow","pink"]
plt.figure(figsize=(6,6))
plt.stackplot(days, studying, playing, Watching_TV, sleeping, labels=labels, colors=colors, alpha=0.4)
plt.legend(loc="upper left")
plt.title("Students Time Distribution")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
```
